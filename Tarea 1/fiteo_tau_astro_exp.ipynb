```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

lndP = np.array([-11.820, -12.187, -12.533, -12.842, -13.101,
                 -13.407, -13.516, -13.779, -14.038, -14.143])
el_angles = np.array([23.5, 16.6, 12.84, 10.48, 8.85,
                 7.66, 6.76, 6.04, 5.47, 4.99])

neg_cosec_el = -1 / np.sin(np.deg2rad(el_angles))

slope, intercept, r_value, p_value, std_err = stats.linregress(neg_cosec_el, lndP)

plt.figure(figsize=(10, 8))

plt.plot(neg_cosec_el, lndP, '-', markersize=8, color='red', label='Datos experimentales')
fit_line = intercept + slope * neg_cosec_el
plt.plot(neg_cosec_el, fit_line, '--', color='darkorange', linewidth=2,
         label=f'Ajuste lineal: $\\tau_0 = {slope:.4f}$\n$R^2 = {r_value**2:.5f}$')

plt.xlabel('$-cosec(el)$', fontsize=12)
plt.ylabel('$\ln(\Delta P)$', fontsize=12)
plt.title('Determinaci√≥n de $\\tau_0$ por Antenna Dipping', fontsize=14)

textstr = '\n'.join((
    f'$\\tau_0 = {slope:.4f}$',))
plt.gca().text(0.05, 0.95, textstr, transform=plt.gca().transAxes,
               fontsize=10, verticalalignment='top', bbox=dict(facecolor='white', alpha=0.8))

plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(loc='lower right', fontsize=10)
plt.tight_layout()

plt.savefig('tau0_fit_simple.png', dpi=300, bbox_inches='tight')
plt.show()
```
